```python
import pandas as pd
import numpy as np
```

```python
db = pd.read_excel("PLANILHA COMPLETA - CLASSIFICAÇÃO.xlsx")
db.info()
```

```python
db.drop(columns=["E_CA", "OBS", "Unnamed: 14", "Unnamed: 15"], axis=1, inplace=True)
db.rename(columns={"L/2": "L2", "D/4": "D4", "L/4": "L4"}, inplace=True)
db.info()
```

```python
db.replace("-", np.nan, inplace=True)
db.replace("    ", np.nan, inplace=True)
db.replace("--", np.nan, inplace=True)

nulos = db[db.isnull().any(axis=1)]
nulos
```

```python
db_clean = db.dropna()
db_clean.info()
```

```python
db_clean = db_clean.astype({"N_OVO": "str", "AL_ALB": "float", "P_GE": "float", "YH": "float", "YD": "float", "YI": "float", "HU": "float", "F_Q": "float", "P_CA": "float"})
db_clean["N_OVO"] = db_clean["N_OVO"].apply(lambda x: 'ovo' + x)
db_clean
```

```python
db_ohe = pd.get_dummies(db_clean, columns=['CL_PESO','RK'])
db_ohe
```

```python
db_ohe.info()
```

```python
db_ohe.to_csv("Medidas todos os ovos.csv", index=False)
```
